```python
import numpy as np

# Function to generate an array of integers
def generate_array(start, end):
    return np.arange(start, end + 1)

# Function to generate an array of linearly spaced values
def generate_linspace(start, end, num_points):
    return np.linspace(start, end, num_points)

# Function to calculate the energy of a signal
def calculate_energy(signal):
    return np.sum(np.abs(signal)**2)

# Function to calculate the trapezoidal integration of a signal
def trapezoidal_integration(signal):
    return np.trapz(signal)

# Discrete values of n for sin[n]
n = generate_array(start=-1000, end=999)
x1 = np.sin(n)
E_x1 = calculate_energy(x1)

# Continuous values of t for cos(t)
t = generate_linspace(start=-1000, end=1000, num_points=10000)
x2 = np.cos(t)
E_x2 = trapezoidal_integration(x2)

# Continuous values of t for 2cos(t) + i*sin(t)
x3 = 2 * np.cos(t) + 1j * np.sin(t)
E_x3 = trapezoidal_integration(x3)

# Discrete values of n for y[n] = (2/sqrt(pi)) * e^(-in)
x4 = (2/np.sqrt(np.pi)) * np.exp(-1j * n)
E_x4 = calculate_energy(x4)

# Display energy values
print("Energy for sin[n]:", E_x1)
print("Energy for cos(t):", E_x2)
print("Energy for 2cos(t) + i*sin(t):", E_x3)
print("Energy for y[n] = (2/sqrt(pi)) * e^(-in):", E_x4)
```

```python
import numpy as np
import matplotlib.pyplot as plt


theta = np.linspace(0, 2*np.pi, 1000)
lhs = np.sin(theta)**2
rhs = 0.5 * (1 - np.cos(2*theta))

plt.plot(theta, lhs, label=r'$\sin^2(\theta)$')
plt.plot(theta, rhs, label=r'$\frac{1}{2}(1 - \cos(2\theta))$')

plt.xlabel(r'$\theta$')
plt.ylabel('Function Value')
plt.legend()

plt.show()
```
